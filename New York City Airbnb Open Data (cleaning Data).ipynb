```python
df_reviews_filtered = df.dropna(subset=['last_review','reviews_per_month'])

print(f"Original number of rows: {len(df)}")
print(f"Number of rows after dropping nulls: {len(df_reviews_filtered)}")
```

```python
df_reviews_filtered['last_review'] = pd.to_datetime(df_reviews_filtered['last_review'], errors='coerce')

print(df_reviews_filtered['last_review'].isnull().sum())
```

```python
df_reviews_filtered['Year_month'] =df_reviews_filtered['last_review'].dt.to_period('M')

df_reviews_filtered.head(7)
```

```python
monthly_data = df_reviews_filtered.groupby('Year_month').agg(
    Monthly_Listings=('id', 'count'),
    Monthly_Reviews=('number_of_reviews', 'sum')
).reset_index()

print(monthly_data)
```

```python
monthly_data['Listings_growth_rate'] = monthly_data['Monthly_Listings'].pct_change()
monthly_data['Reviews_growth_rate'] = monthly_data['Monthly_Reviews'].pct_change()

monthly_data.loc[0, 'Listings_Growth_Rate'] = None
monthly_data.loc[0, 'Reviews_Growth_Rate'] = None

print(monthly_data[['Year_month', 'Listings_growth_rate', 'Reviews_growth_rate']])
```

```python
df_reviews_filtered = pd.merge(df_reviews_filtered, monthly_data, on='Year_month', how='left')

df_reviews_filtered.head(7)
```

```python
df_reviews_filtered = df_reviews_filtered.drop(columns=['Listings_Growth_Rate', 'Reviews_Growth_Rate'])

df_reviews_filtered.head(7)
```

```python
df_reviews_filtered['Year_month'] = pd.to_datetime(df_reviews_filtered['Year_month'].astype(str))
```

```python
plt.figure(figsize=(10, 6))
for group, data in df_reviews_filtered.groupby('neighbourhood_group'):
    plt.plot(data['Year_month'], data['Listings_growth_rate'], label=group)

plt.title('Listings Growth Rate by Neighbourhood Group')
plt.xlabel('Year-Month')
plt.ylabel('Listings Growth Rate')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
